```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm= ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.1
)
```

```python
# here we are augmenting structured output for response we wil get
```

```python
from pydantic import BaseModel , Field

class SearchQuery(BaseModel):
    search_query:str=Field(None,description="Query that is opimized web search")
    justification: str=Field(None,description="Why this query is relevant to the user's request")
```

```python
#augment the LLM with the schema for structured output

structured_llm=llm.with_structured_output(SearchQuery)

#invoke the augmented llm
output=structured_llm.invoke("How does Calcium CT score relate to high cholesterol?")

output
```

```python
#define a tool

def multiply(a:int,b:int)->int:
    return a*b

    
```

```python
#augemnt llm with tools

llm_with_tools=llm.bind_tools([multiply])
```

```python
#invoke the LLM with input that triggers the call

msg=llm_with_tools.invoke("What is 2 times 3")
```
